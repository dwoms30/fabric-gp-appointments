```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
from pyspark.sql.functions import col, to_date, when

#read bronze ref csv
df_ref = spark.table("bronze.practice_reference_raw")

#silver rules applied
df_ref_silver = (
    df_ref
    .dropDuplicates(["practice_code"])
    .filter(col("practice_code").isNotNull())
)

#write silver table
df_ref_silver.write.mode("overwrite").saveAsTable("silver.practice_reference_clean")

```
