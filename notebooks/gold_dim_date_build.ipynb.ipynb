```python
#creating date dimension (data driven instead of conformed as lightweight project)
from pyspark.sql.functions import col, year, month, quarter, weekofyear, date_format

df_dates = (
    spark.table("silver.appointments_clean")
    .select("appointment_date")
    .distinct()
    #creating date key
    .withColumn("date_key", date_format("appointment_date", "yyyyMMdd").cast("int"))
    .withColumn("year", year("appointment_date"))
    .withColumn("month", month("appointment_date"))
    .withColumn("quarter", quarter("appointment_date"))
    .withColumn("week_of_year", weekofyear("appointment_date"))
    .withColumn("financial_year",
        year("appointment_date") +
        (month("appointment_date") >= 4).cast("int")
    )
)

df_dates.write.mode("overwrite").saveAsTable("gold.dim_date")


```
