```python
# Building the gold practice dimensions table
from pyspark.sql.functions import monotonically_increasing_id

df_practice = (
    spark.table("silver.practice_reference_clean")
    .select(
        "practice_code",
        "practice_name",
        "borough",
        "pcn_name"
    )
    .distinct()
    # Creating surrogate key for practice
    .withColumn("practice_key", monotonically_increasing_id())
)




df_practice.write.mode("overwrite").saveAsTable("gold.dim_practice")
```
