```python
# Read reference data from Bronze files

df_practice = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "false")
    .csv("Files/bronze/reference/practice_reference.csv") 
)

#Minimal shaping (Power Query equivalent)

df_practice_clean = (
    df_practice
    .select(
        "practice_code",
        "practice_name",
        "pcn_name",
        "borough",
        "pcn_code",
        "region_name",
        "region_code"
    )
)

df_practice_clean.write.mode("overwrite").saveAsTable("bronze.practice_reference_raw")
```
